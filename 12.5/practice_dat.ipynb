8. Create a DataFrame of 8 students with columns "Name" and "Score". 

Filter and print students with Score > 60 and those with Score < 40.

```python
st_data = {
  'Name' : [
    'Hijbullah', 'Rony', 'Sazid', 'Kamal', 'Rahim', 'Avir', 'Arif', 'Soikat'
  ],
  'Score' : [
    100,99,23,44,60,80,55,80
  ]
}

s_df = pd.DataFrame(st_data)
s_df
```

```python
filtered_st = s_df.loc[(s_df['Score'] > 60) | (s_df['Score'] < 40)]
print(filtered_st)
```
